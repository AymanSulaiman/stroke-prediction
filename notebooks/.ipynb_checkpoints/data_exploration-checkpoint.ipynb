```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(os.path.join('..','data','healthcare-dataset-stroke-data.csv'))
df
```

```python
# Looking at the unique categorical data
print(f"Genders: {df.gender.unique()}")
print(f"Marriage Status: {df.ever_married.unique()}")
print(f"Working Type: {df.work_type.unique()}")
print(f"Residence Type: {df.Residence_type.unique()}")
print(f"Smoking Status: {df.smoking_status.unique()}")
```

```python
plt.hist(df.age)
plt.xlabel('Age');
plt.ylabel('Count');
plt.title('Age Histogram')
plt.show();
```

```python
def get_scatter(x_axis, y_axis):
    x_axis_lower = x_axis.lower()
    y_axis_lower = y_axis.lower()
    
    plt.scatter(x=df[df.gender == 'Male'][x_axis_lower], y=df[df.gender == 'Male'][y_axis_lower], color='blue', alpha=0.4, edgecolors='black', label='Male');
    plt.scatter(x=df[df.gender == 'Female'][x_axis_lower], y=df[df.gender == 'Female'][y_axis_lower], color='red', alpha=0.4, edgecolors='black', label='Female');
    plt.scatter(x=df[df.gender == 'Other'][x_axis_lower], y=df[df.gender == 'Other'][y_axis_lower], color='green', alpha=0.4, edgecolors='black', label='Other');
    plt.xlabel(x_axis);
    plt.ylabel('BMI');
    plt.legend();
    plt.title(f'{x_axis} against {y_axis}');
    plt.figure(figsize=(35, 25));
    plt.show();
    
get_scatter("Age", "BMI")
```
